```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np
from pandas.io.parsers import read_csv
from sklearn.preprocessing import StandardScaler
```

```python
# 데이터 로드
data = read_csv('price data.csv', sep=',')

# 'year' 열을 제외한 나머지 열을 사용하여 데이터 배열 생성
xy = data[['avgTemp', 'minTemp', 'maxTemp', 'rainFall', 'avgPrice']].to_numpy()

# 입력 데이터(x_data)와 출력 데이터(y_data) 분리
x_data = xy[:, :-1]  # 평균 기온, 최저 기온, 최고 기온, 강수량
y_data = xy[:, [-1]]  # 평균 가격
```

```python
# 데이터 정규화: 입력 데이터를 표준화하여 학습 성능 향상
scaler = StandardScaler()
x_data = scaler.fit_transform(x_data)

# TensorFlow 플레이스홀더 정의
X = tf.placeholder(tf.float32, shape=[None, 4])  # 입력 데이터: 4개의 특성
Y = tf.placeholder(tf.float32, shape=[None, 1])  # 출력 데이터: 1개의 레이블 (가격)
```

```python
# 가중치와 편향 초기화
W = tf.Variable(tf.random_normal([4, 1]), name="weight")  # 가중치: 4개의 입력과 1개의 출력
b = tf.Variable(tf.random_normal([1]), name="bias")  # 편향: 1개의 출력

# 가설 정의
hypothesis = tf.matmul(X, W) + b

# 비용 함수 정의: 평균 제곱 오차 (MSE)
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# 옵티마이저 정의: 경사 하강법 사용, 학습률 0.0001
learning_rate = 1e-4
optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
train = optimizer.minimize(cost)

# TensorFlow 세션 실행
sess = tf.Session()
sess.run(tf.global_variables_initializer())  # 변수 초기화
```

```python
# 모델 학습
for step in range(100001):  # 100,000번 반복
    cost_, hypo_, _ = sess.run([cost, hypothesis, train], feed_dict={X: x_data, Y: y_data})  # 학습 단계
    if step % 500 == 0:  # 500번마다 손실 비용과 예측 값 출력
        print("#", step, " 손실 비용: ", cost_)
        print("-사과 가격: ", hypo_[0])

# 학습된 모델 저장
saver = tf.train.Saver()
save_path = saver.save(sess, "./saved.cpkt")
print("학습된 모델을 저장했습니다.")
```
